```python
import numpy as np

def trimf(x, a, b, c):
    mu = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] <= a:
            mu[i] = 0
        elif x[i] <= b:
            mu[i] = (x[i] - a) / (b - a)
        elif x[i] <= c:
            mu[i] = (c - x[i]) / (c - b)
        else:
            mu[i] = 0
    return mu

def trapmf(x, a, b, c, d):
    mu = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] <= a:
            mu[i] = 0
        elif x[i] <= b:
            mu[i] = (x[i] - a) / (b - a)
        elif x[i] <= c:
            mu[i] = 1
        elif x[i] <= d:
            mu[i] = (d - x[i]) / (d - c)
        else:
            mu[i] = 0
    return mu

def gaussmf(x, c, sigma):
    mu = np.exp(-1 / 2 * ((x - c) / sigma) ** 2)
    return mu

def gbellmf(x, a, b, c):
    mu = 1 / (1 + abs((x - c) / a) ** (2 * b))
    return mu

def sigmf(x, a, c):
    mu = 1 / (1 + np.exp(-a * (x - c)))
    return mu
```

```python
import matplotlib.pyplot as plt
```

```python
start = 0
stop = 10
step = 0.2
num = int(stop / step + 1)
x = np.linspace(start, stop, num)

#pertinencias do plot superior
a_1 = 3
b_1 = 4
c_1 = 5
mu_1 = trimf(x, a_1, b_1, c_1)
a_2 = 2
b_2 = 4
c_2 = 7
mu_2 = trimf(x, a_2, b_2, c_2)
a_3 = 1
b_3 = 4
c_3 = 9
mu_3 = trimf(x, a_3, b_3, c_3)
#pertinencias do plot inferior
a_4 = 2
b_4 = 3
c_4 = 5
mu_4 = trimf(x, a_4, b_4, c_4)
a_5 = 3
b_5 = 4
c_5 = 7
mu_5 = trimf(x, a_5, b_5, c_5)
a_6 = 4
b_6 = 5
c_6 = 9
mu_6 = trimf(x, a_6, b_6, c_6)

#plot
fig, axs = plt.subplots(2)
axs[0].plot(x, mu_1, color='b')
axs[0].plot(x, mu_2, color='g')
axs[0].plot(x, mu_3, color='r')
axs[1].plot(x, mu_4, color='b')
axs[1].plot(x, mu_5, color='g')
axs[1].plot(x, mu_6, color='r')
plt.setp(axs, xticks=np.linspace(start=0, stop=10, num=11),
         yticks=np.linspace(start=0, stop=1, num=6));
```
